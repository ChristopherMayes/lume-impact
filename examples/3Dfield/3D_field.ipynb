## Use Cartesian 3D field in lume-Impact

```python
from impact import Impact
import os

# plotting modules
from impact.plotting import layout_plot
from bokeh.plotting import output_notebook
from bokeh.plotting import show
output_notebook(hide_banner=True)
```

```python
# Make Impact object
ifile = 'ImpactT.in'
I = Impact(ifile, verbose=True)
```

```python
# Show a MAD-style printout of each element in the lattice
I.print_lattice()
```

```python
I.plot(figsize=(9,6))
```

```python
layout = layout_plot(I.input['lattice'], plot_height=300, plot_width=600)
show(layout)
```

```python
I.run()
```

```python
I.plot(figsize=(9,6))
```
